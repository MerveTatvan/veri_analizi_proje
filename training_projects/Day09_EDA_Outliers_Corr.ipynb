```python
# IQR yÃ¶ntemi ile outlier tespiti
outlier_dict = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_dict[col] = outliers.shape[0]
    print(f"{col} kolonunda outlier sayÄ±sÄ±: {outliers.shape[0]}")
```

```python
# TÃ¼m sayÄ±sal kolonlar iÃ§in outlier'larÄ± filtreleyip yeni DataFrame oluÅŸturalÄ±m
df_filtered = df.copy()

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df_filtered = df_filtered[(df_filtered[col] >= lower_bound) & (df_filtered[col] <= upper_bound)]

print(f"Filtrelendikten sonra DataFrame boyutu: {df_filtered.shape}")
```

```python
# Correlation matrix hesaplayalÄ±m
corr_matrix = df_filtered[numeric_cols].corr()
print(corr_matrix)

# Heatmap ile gÃ¶rselleÅŸtirme
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
